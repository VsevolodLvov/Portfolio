# 1. Импорт датасета и преобразование данных.

Ниже приведён мой код, для загрузки данных. В зависимости от формата и расположения файла данных, код будет иным.

**Загрузка данных.**

```python
# My Code:
# Загружаем файл с данными
path = 'data/telecom_users.csv'  # Определяем путь к файлу

df_main = pd.read_csv(
    path,
    sep = ',', 
    header = 0, 
    index_col = 0
)

# # Your Code:
# df_main = 
```

**Преобразование типов данных для совместимости с основной работой.**

```python
# ДЛЯ СОВМЕСТИМОСТИ
# Повторим здесь код преобразований сделанных в основной работе для целей анализа данных. 
# С точки зрения последующей бинаризации и дискретизации он излишний.

# Изменим типы значений отедльных столбцов для соотвествия смыслу данных
# SeniorCitizen содержит категории
df_main.SeniorCitizen = df_main.SeniorCitizen.apply(lambda x: 'Yes' if x == 1 else "No")

# TotalCharges - должен быть числового типа.
df_main.TotalCharges = pd.to_numeric(
    df_main.TotalCharges, errors='coerce', downcast=None) # недопустимые занечения заменим на NaN (coerce)

# Заменим NaN на 0
df_main = df_main.fillna(0)
print(f"Осталось NaN: {pd.isna(df_main).sum().sum()}")
```

**Дискретизация.**

```python
# Словарь дискретайзеров
dict_discretizers = joblib.load('data/dict_discretizers.joblib')

# Используем все данные df_main, кроме первого столбца, содержащего уникальные ID клиентов
df_wrk = df_main.iloc[:,1:]

# Определим список столбцов с числовыми данными (с большим количеством значений или непрерывными)
list_col_numerals = (
    df_main
    .dtypes  # получаем серию, где индексом являются названия столбцов, значениями - типы данных
    .loc[df_main.dtypes != 'object']  # выбираем тип данных НЕ'object'
    .index  # получаем индекс
    .tolist()  # преобразуем в список.
)

# В цикле произведём дискретезацию и заменим значения в числовых столцах df_wrk
for col in list_col_numerals:
    df_wrk.update(
        pd.Series(
            (dict_discretizers[col].fit_transform(np.array(df_wrk[col]).reshape(-1, 1)))
            .reshape(1,-1)[0],
            name=col,
            index=df_wrk.index,
            dtype='Int32'),
        overwrite=True
    )
```

**Бинаризация.**

```python
# Сформмруем список категорий для OneHotEncoder
list_extras = ['No phone service', 'No internet service']  # Список излишних категорий
list_list_cat = []  # Инициализируем список списков нужных категорий
for col in df_wrk.keys():  # В цикле по названиям столбцов
    list_cat=[]  # Инициализируем список нужных категорий
    for val in df_wrk.loc[:,col].unique():  # В цикле по уникальным значения данного столбца
        if val in list_extras:  # Если значение входит в список излишних категорий
            continue  # Пропускаем итерацию
        else:
            list_cat.append(val)  # Иначе, добавляем значение в список нужных категорий
    list_cat.sort()  # Сортируем список нужных категорий
    list_list_cat.append(list_cat)  # Добавляем его в список списков
# list_list_cat    

# Создадим объект кодирования one-hot для обеспечения бинарности категорий 
enc = OneHotEncoder(categories=list_list_cat, # передаём список списков нужных категорий, по стобцам DF
                    drop='if_binary',  # Используем только одну категорию, если исходный столбец бинарный
                    sparse_output=False, 
                    dtype='int32', 
                    handle_unknown='ignore',  # Игнорируем значения категорий, не входящих в список
                    min_frequency=None, 
                    max_categories=None)
# Преобразуем в значения int только категориальные столбцы
data_cat = enc.fit_transform(df_wrk)  # Трансформируем DF

# Обновим рабочий датафрейм со значениями категорий, выраженными через int
df_wrk = pd.DataFrame(
    data_cat, 
    columns=enc.get_feature_names_out(), 
    index=df_wrk.index
)
```

Полученное выше предупреждение вполне ожидаемо: мы передали кодировщику данные, в которых есть значения категорий, отсутствующие в списке категорий, предназначенных для кодирования. Благодаря handle_unknown='ignore' эти категории будут закодированы нулями, а столбцы с ними не войдут в вывод метода get_feature_names_out().

**Формирование массивов X и y.**

**Функции для выборки признаков из датасета по их весам (для работы с FunctionTransformer).**

```python
def ranked_features_extract(X_=None, y_=None, ser_feat_=None, limit_coef_=0):
    """
    Принимает 
    - X, [y],
    - отсортированную серию признаков с весами (коэффициентами корреляции с y, долями в структуре y и т.п.)
    - минимальное абсолютное значение веса признаков с для извлечения (если limit_coef_<=1) 
    или количество признаков для извлечения из X с наибольшими весами (если limit_coef_>1), 
    в последнем случае limit_coef_ будет округлён до ближайшего целого числа.
    Возвращает 
    - df признаков Х, соответствующих критерию весов: 
    не ниже абсолютного значения веса: >= limit_coef_ 
    (если limit_coef_<=1), либо соответствующих заданному количеству признаков,
    ранжированных по весам (если limit_coef_>1)

    """
   
    if limit_coef_ <=1:
        # Создаём список параметров, с абсолютным уровнем веса на уровне limit_coef_ и вывше.
        list_feat_ = ser_feat_[ser_feat_ >= limit_coef_].index.tolist()  
    else:
        # Создаём список из limit_coef_ числа параметров, ранжированных по весам, если limit_coef_>1.
        list_feat_ = ser_feat_.nlargest(round(limit_coef_)).index.tolist()
        
    X_tr_=X_[list_feat_]
#     y_tr_ = y_
    return np.array(X_tr_)

def transformed_ranked_feature_names(transformer_, input_features):
    """
    Принимает экземпляр трансформера, и массив входных признаков (см. мануал sklearn FunctionTransformer)
    Возвращает список отобранных признаков
    """
    
    ser_feat_ = transformer_.kw_args['ser_feat_']
    limit_coef_ = transformer_.kw_args['limit_coef_']
    
    # Создаём список параметров, с абсолютным уровнем корреляции с целевой переменной на уровне limit_coef_ и вывше.
   
    if limit_coef_ <=1:
        # Создаём список параметров, с абсолютным уровнем веса на уровне limit_coef_ и вывше.
        list_feat_ = ser_feat_[ser_feat_ >= limit_coef_].index.tolist()  
    else:
        # Создаём список из limit_coef_ числа параметров, ранжированных по весам, если limit_coef_>1.
        list_feat_ = ser_feat_.nlargest(round(limit_coef_)).index.tolist()       
    return list_feat_
    
```
